```python
import os
import random
import pandas as pd
```

```python
c = '/workspaces/classify_covid/data/raw/COVID/images'
n = '/workspaces/classify_covid/data/raw/Normal/images'
p = '/workspaces/classify_covid/data/raw/Viral Pneumonia/images'

random.seed(42)
filenames = os.listdir(c) + random.sample(os.listdir(n), 5000) + os.listdir(p)
```

```python
filenames
```

```python
def create_df(filenames, c, p, n):
  categories = []
  for filename in filenames:
    category = filename.split('-')[0]
    print(category)
    if category == 'COVID':
      categories.append(str(2))
    elif category == 'Viral Pneumonia':
      categories.append(str(1))
    else:
      categories.append(str(0))

  """converting to dataframe"""
  for i in range(len(filenames)):
    if 'COVID' in filenames[i]:
      filenames[i] = os.path.join(c, filenames[i])
    elif 'Viral Pneumonia' in filenames[i]:
      filenames[i] = os.path.join(p, filenames[i])
    else:
      filenames[i] = os.path.join(n, filenames[i])

  df = pd.DataFrame({
    'filename': filenames,
    'category': categories
  })
  return df
```

```python
df = create_df(filenames, c, p, n)
```
